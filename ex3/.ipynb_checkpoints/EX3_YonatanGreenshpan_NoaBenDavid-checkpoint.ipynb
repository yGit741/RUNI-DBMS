```python
# Initializae function that will help us to generate valid data

def generate_unique_ids(n, existing_ids=False):
    ids = set()
    if existing_ids:
        existing_ids_set = set(existing_ids)
        while len(ids) < n:
            id_number = str(random.randint(100000000, 999999999))
            if id_number not in existing_ids_set:
                ids.add(id_number)
    else:
        while len(ids) < n:
            id_number = str(random.randint(100000000, 999999999))
            ids.add(id_number)
    return list(ids)

def check_above_18(date_of_birth):
    # Convert the date of birth string to a datetime object
    dob = datetime.datetime.strptime(date_of_birth, '%Y-%m-%d')
    
    # Calculate the age by subtracting the date of birth from the current date
    age = datetime.datetime.now().year - dob.year
    
    # Check if the age is above 18
    if age > 18:
        return True
    else:
        return False
    

def generate_random_full_name():
    first_name = random.choice(first_names)
    last_name = random.choice(last_names)
    full_name = f"{first_name} {last_name}"
    return full_name

def generate_address():
    street = random.choice(street_names)
    number = random.choice(list(range(1, 50)))
    return f"Tel-Aviv, {street} {number}"

```

```python
# A function that delete table if exist that help run cells of creating and inserting data without reset the kernel

def delete_table_if_exists(table_name):
    c.execute("SELECT name FROM sqlite_master WHERE type='table' AND name=?", (table_name,))
    result = c.fetchone()

    if result is not None:
        c.execute("DROP TABLE " + table_name)
        print("Table '{}' deleted successfully.".format(table_name))
    else:
        print("Table '{}' does not exist.".format(table_name))

    # Commit the changes and close the connection
    conn.commit()
```

```python
delete_table_if_exists('Orders')

c.execute('''
    CREATE TABLE IF NOT EXISTS "Orders" (
        CustomerID INTEGER,
        TimeDate TEXT,
        CatalogNumber INTEGER,
        DroneID INTEGER,
        PRIMARY KEY (CatalogNumber, CustomerID, TimeDate),
        FOREIGN KEY (CustomerID) REFERENCES Customers(ID),
        FOREIGN KEY (CatalogNumber) REFERENCES Products(CatalogNumber),
        FOREIGN KEY (DroneID) REFERENCES Drones(DroneID)
    )
''')

c.execute('''
SELECT CatalogNumber FROM Foods
''') 
Foods_CatNum = [CatNum[0] for CatNum in c.fetchall()]


c.execute('''
SELECT CatalogNumber FROM Flowers
''') 
Flowers_CatNum = [CatNum[0] for CatNum in c.fetchall()]

# Generate random orders for each customer
c.execute("SELECT ID FROM Customers")
customer_ids = [row[0] for row in c.fetchall()]
CatNum = Foods_CatNum + Flowers_CatNum
DroneID = list(range(1,11))

for customer_id in customer_ids:
    # Generate a random number of orders (between 1 and 10)
    num_orders = random.randint(1, 10)

    for _ in range(num_orders):
        # Generate a random catalog number
        catalog_number = random.choice(CatNum)

        # Generate a random date within the past two years
        today = datetime.date.today()
        two_years_ago = today - datetime.timedelta(days=2 * 365)
        random_date = random.choice(
            [two_years_ago + datetime.timedelta(days=x) for x in range((today - two_years_ago).days)]
        )

        # Generate a random time within the day (normal distribution with mu=12, scale=2)
        random_hour = int(random.gauss(12, 2))
        random_minute = random.randint(0, 59)
        random_time = datetime.time(random_hour, random_minute)

        # Generate the combined primary key (CustomerID, TimeDate, CatalogNum)
        primary_key = (customer_id, str(random_date) + ' ' + str(random_time), catalog_number, random.choice(DroneID))

        # Insert the order into the Orders table
        c.execute("INSERT INTO Orders (CustomerID, TimeDate, CatalogNumber, DroneID) VALUES (?, ?, ?, ?)", primary_key)

# Commit the changes and close the connection

c.execute(''' 
SELECT * FROM Orders
''')
conn.commit
pd.DataFrame(c.fetchall(), columns=['CustomerID', 'TimeDate', 'CatalogNumber', 'DroneID'])
```

## Exploring the Data

Let's have a look on the following queries

Finding a problematic drone and an outstanding drone

```python

c.execute('''
SELECT Drones.*, COUNT(*) AS TotalQuantitySold
FROM Orders
JOIN Drones ON Orders.DroneID = Drones.DroneID
GROUP BY Drones.DroneID
ORDER BY TotalQuantitySold DESC
LIMIT 1;
''')

pd.DataFrame(c.fetchall(), columns=['DroneID', 'Manufacturer', 'Model', 'PayloadCapacity', 'AlgoUses', 'TotalQuantitySold'])
```

```python

c.execute('''
SELECT Drones.*, COUNT(*) AS TotalQuantitySold
FROM Orders
JOIN Drones ON Orders.DroneID = Drones.DroneID
GROUP BY Drones.DroneID
ORDER BY TotalQuantitySold
LIMIT 1;
''')

pd.DataFrame(c.fetchall(), columns=['DroneID', 'Manufacturer', 'Model', 'PayloadCapacity', 'AlgoUses', 'TotalQuantitySold'])
```

Comparing the too algorithms

```python

c.execute('''
SELECT D.AlgoUses, SUM(P.Price) AS Revenue
FROM Drones D
JOIN Orders O ON D.DroneID = O.DroneID
JOIN Products P ON O.CatalogNumber = P.CatalogNumber
GROUP BY D.AlgoUses;
''')

pd.DataFrame(c.fetchall(), columns=['Algorithm', 'Revenue'])
```

Who order more food and flowers men or women?

```python


c.execute('''
SELECT c.Gender, p.Type, COUNT(*) AS OrderCount
FROM Customers c
JOIN Orders o ON c.ID = o.CustomerID
JOIN Products p ON o.CatalogNumber = p.CatalogNumber
GROUP BY c.Gender, p.Type
''')

pd.DataFrame(c.fetchall())
```

Here we will show how many of the costumers are eligible to get discount, whcih mean that their partner made an order.

```python
c.execute('''
SELECT c1.ID, c1.FullName, c1.Gender, c1.Address, c1.DateOfBirth, c1.PartnerID, c2.FullName AS PartnerName
FROM Customers AS c1
JOIN Customers AS c2 ON c1.PartnerID = c2.ID
WHERE c1.ID IS NOT NULL;
''')

pd.DataFrame(c.fetchall(), columns=['ID', 'FullName', 'Gender', 'Address', 'DateOfBirth', 'PartnerID', 'PartnerName'])
```

Find how many orders have been made to Allenby street?

```python
c.execute('''
SELECT COUNT(*) AS OrderCount
FROM Orders AS o
JOIN Customers AS c ON o.CustomerID = c.ID
WHERE c.Address LIKE '%Ben Yehuda%';
''')

pd.DataFrame(c.fetchall())
```

What is the Food Buisness branch that is the most popular among costumers?

```python
c.execute('''
SELECT f.Name, COUNT(o.TimeDate) AS Popularity
FROM Foods AS f
JOIN Orders AS o ON f.CatalogNumber = o.CatalogNumber
GROUP BY f.Name
ORDER BY Popularity DESC;

''')

pd.DataFrame(c.fetchall())
```

## Bonuse Exploration

Here we will use th COALESCE function that we didn't learn in class. We aim to answer the question what are the "three products that gave the highes revenue?" (without this function we will get two columns with NULLs because the name is on two differenct tables: Foods and Flowers).

```python
c.execute('''
SELECT O.CatalogNumber, COALESCE(F.Name, FD.Name) AS ProductName, P.Price * COUNT(*) AS TotalRevenue
FROM Orders O
JOIN Products P ON O.CatalogNumber = P.CatalogNumber
LEFT JOIN Flowers F ON O.CatalogNumber = F.CatalogNumber
LEFT JOIN Foods FD ON O.CatalogNumber = FD.CatalogNumber
GROUP BY O.CatalogNumber
ORDER BY TotalRevenue DESC
LIMIT 3
''')

pd.DataFrame(c.fetchall())
```

Here we will use th LAG function that we didn't learn in class. We aim to check the average gap between time of orders grouped by the number of orders

```python

c.execute('''

SELECT OrderCount, AVG(julianday(TimeDate) - julianday(LagTime)) AS MeanOrderGapInDays
FROM (
    SELECT CustomerID, TimeDate, LAG(TimeDate) OVER (PARTITION BY CustomerID ORDER BY TimeDate) AS LagTime,
           COUNT(*) OVER (PARTITION BY CustomerID) AS OrderCount
    FROM Orders
) AS Subquery
GROUP BY OrderCount;
''')

pd.DataFrame(c.fetchall(), columns=['Number of Orders','Mean Order Gap In Days'])
```
